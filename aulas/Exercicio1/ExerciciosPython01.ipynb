```python
#  Questão 10

med1 = int(input('Coloque o primeiro tamanho do triângulo: '))
med2 = int(input('Coloque o segundo tamanho do triângulo: '))
med3 = int(input('Coloque o terceiro tamanho do triângulo: '))
 
if((med1 + med2 > med3) and (med2 + med3 > med1) and (med1 + med3 > med2)): 
  if(med1 == med2 and med1 == med3 and med2 == med3):
    print('O triângulo é Equilatero.')
  elif(med1 == med2 or med1 == med3 or med2 == med3):
    print('O triângulo é Isósceles.')
  else:
    print('O triângulo é Escaleno.')
else:
  print('Não é um triângulo')
```

```python
#  Questão 11

angulo1 = int(input('Coloque o primeiro valor do ângulo do triângulo: '))
angulo2 = int(input('Coloque o segundo valor do ângulo do triângulo: '))
angulo3 = int(input('Coloque o terceiro valor do ângulo do triângulo: '))

if(angulo1 == 90 or angulo2 == 90 or angulo3 == 90):
  print('É um triângulo retângulo, pois possui um ângulo reto')
elif(angulo1 > 90 or angulo2 > 90 or angulo3 > 90):
  print('É um triângulo Obtusângulo, pois possui um ângulo maior que 90°')
elif(angulo1 < 90 and angulo2 < 90 and angulo3 < 90):
  print('É um triângulo Acutângulo, pois possui três ângulos menor que 90°')
else:
  print('Valor não aceito, coloque valores válidos para achar os três triângulos válidos')

```

```python
#  Questão 12

combustivel = int(input('Coloque aqui o combustível você usou para abastecer "1" para álcool ou "2" para Gasolina : '))

if(combustivel == 1):  
  a = int(input('Coloque a quantidade que foi abastecida de álcool: '))
  if(a <= 20):
    valorA = a * 0.03  
    print('O valor a ser pago é de: R$', a - valorA)
  elif(a > 20):
    valorB = a * 0.05
    print('O valor a ser é de: R$', a - valorB)
  else: 
    print('Coloque apenas números.')
elif(combustivel == 2):
  g = int(input('Coloque a quantidade que foi abastecida de gasolina: '))
  if(g <= 20):
    valorA = g * 0.04
    print('O valor a ser pago é de: R$', g - valorA)
  elif(g > 20):
    valorB = g * 0.06
    print('O valor a ser é de: R$', g - valorB)
  else:
    print('Coloque apenas números.')
else: 
  print('Apenas os valores "1" e "2" serão válidos, por favor use-os apenas!')
  

```
