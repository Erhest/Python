```python
# Faça um programa que leia um nome de usuário e a sua senha e não aceite a senha igual ao nome do usuário
#  mostrando uma mensagem de erro e voltando a pedir as informações.

continuar = True
while(continuar): 
  usuario = input('Digite seu USUÁRIO: ')
  senha = input('Digite a sua SENHA: ')
  if(usuario != senha):
    continuar = False
  else: 
    print('Digite valores diferentes')
```

```python
#  Desenvolva uma calculadora ccom menu para selecionar o tipo de 
#  conta adição, subtração, multiplicação e divisão

continuar = True

while(continuar):
  menu = input('**** Calculadora **** \n 1 - Adição \n 2 - Subtração \n 3 - Multiplicação \n 4 - Divisão \n Digite qualquer outro valor para sair \n')
  if(menu.isnumeric):
    menu = int(menu)
    if(menu > 0 and menu < 5):
      num1 = int(input('Digite um número: '))
      num2 = int(input('Digite outro número: '))
      if(menu == 1):
        print(num1 + num2)
      elif(menu == 2):
        print(num1 - num2)
      elif(menu == 3):
        print(num1 * num2)
      elif(menu == 4):
        print(num1 / num2)
    else:
      continuar = False
  else: 
    continuar = False
```

```python
#  Faça um programa que leia 5 números e informe a soma e a média dos números.

soma = 0
contador = 0
while(contador < 5):
  num = int(input('Digite um número: '))
  soma += num
  contador+= 1 # contador = contador + 1
print('Total', soma)
print('Média', soma / contador)
```

```python
#  Faça um jogo da sorte que guarde um número (0 - 99) e peça poara o usuário
#  tentar acertar o numero, exiba se errou pra mais ou menos
from random import randint

segredo = randint(0, 99)
continuar = True
contador = 1

while(continuar):
  chute = int(input('Digite um valor entre 0 e 99: \n'))
  if(chute == segredo):
    print('Você acertou, Miseravi!', segredo)
    print('Você usou {} chances! ' .format(contador))
    continuar = False
  elif(chute > segredo):
    print('Menos')
  else:
    print('Mais')
  contador += 1
```
