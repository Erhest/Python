#Vetores

## TUPLA
### VARIAVEL COMPOSTA

#### AS TUPLAS SÃO IMUTÁVEIS => NÃO PODEM SER ALTERADAS APÓS A DECLARAÇÃO 

##### DECLARAÇÃO 
##### nomeTupla = (valor, valor, valor, valor, valor)

```python
# Exemplo

linguagens = ('C', 'Java', 'JavaScript', 'C++', 'C#', 'PHP', 'Python', 'Ruby', 'R') # Criando uma TUPLA
print(type(linguagens))

linguas = ('C') # Variável
print(type(linguas))

ling = tuple('C') # Tupla
print(type(ling))

lin = tuple('C',) # Tupla
print(type(lin))

print(linguagens)
print(linguagens[2])
print(linguagens[5])
print(linguagens[7])

print(len(linguagens)) # O tamanho da Tupla / Vetor

print(linguagens[-1]) # Ultimo valor
print(linguagens[len(linguagens) - 1]) # Ultimo valor
```

```python
# Fatiamemto
## Linguagens

print(linguagens[1:5]) # 1 - 4
print(linguagens[2:])  # 2 - ultimo
print(linguagens[:4]) # 4 primeiros
print(linguagens[-3]) # 3 de trás para frente
print(linguagens[-4:]) # 4 ultimos
```

```python
# Alterando TUPLA
linguagens[4] = 'Go' # Tuplas são imutáveis
print(linguagens)
```

```python
## Funções Nativas das Tuplas

num = (10, 20, 15, -5)

# Ordenar
print(linguagens)
print(sorted(linguagens)) #sorted(tupla) -> não altera tupla
print(linguagens)

print(num)
print(sorted(num))
print(num)

# Buscar pelo valor
print(linguagens.index('Java'))
print(num.index(-5))

# máximo
print(max(num))

# minimo
print(min(num))

# Soma
print(sum(num))

# Tamanho
print(len(num))

# Deletar
del (num)
print(num)
```

# LISTA
## VARIAVEL COMPOSTA
###DECLARAÇÃO
#### nomeTupla=[valor, valor, valor, valor]

```python
# Criando listas
roupas = ['Blusa', 'Calça', 'Meia', 'Casaco'] # Criando Lista
print(type(roupas)) # Tipo
print(roupas) # Exibindo todos os valores da lista
print(len(roupas)) # Tamanho da Lista

numeros = list(range(10, 31, 3))
print(type(numeros))
print(numeros)
print(len(numeros))

teste = list()
print(type(teste))
print(teste)
print(len(teste))
```

```python
# Declarando
carros = ['Palio', 'Gol', 'Celta', 'Corsa']
print(carros)
print(len(carros))

# Adicionar um novo item
carros.append('Fusca') # Adiciona ao final da lista --> ultimo item
# carros.append(input('Adicione um novo veiculo: '))
print(carros)
print(len(carros))

# Inserir em um indice especifico
carros.insert(2, 'Fox')
# carros.insert(input('Digite um carro: '))  // carros.insert(2, variavel)
print(carros)
print(len(carros))

# Remover pelo indice
del carros[2]
print(carros)
print(len(carros))

carros.pop(-1) # Removendo o ultimo item
print(carros)
print(len(carros))

# Remover pelo valor
carros.remove('celta') 
print(carros)
print(len(carros))

# Alterar pelo valor
carros[0] = 'Celta'
print(carros)
print(len(carros))

carros[1] = 'Palio'
print(carros)
print(len(carros))
```

```python
# Copia de tabela
carros = ['Palio', 'Gol', 'Celta', 'Corsa']
outrosCarros = carros[:]
print(outrosCarros)
print(len(outrosCarros))
```

```python
# Teste Lógico

#  Existe o valor TAAL dentro da LISTA
car = input('Pesquise o carro da loja: ')
if(car in carros):
  print('Tem')
else:
  print('Não')
```

```python
# Funções
# Ordenar
valores= [15, 20, 98, -59, 275, 159.2, 2]
print(valores)
print(len(valores))
valores.sort() # Altera Lista
print(valores)
valores.sort(reverse=True) # Decrencente (precisa estar junto o código reverse=True)
print(valores)

# Max
print(max(valores))

# Min
print(min(valores))

# Tamanho
print(len(valores))

# Soma
print(sum(valores))

# Ordenar sem alterar a lista
print(sorted(valores)) # Não Altera a Lista
print(valores)
```

```python
# Listando itens dos vetores
alunos = ['Kenji', 'Mateus', 'Akio', 'Emerson', 'Davi', 'Sayumi', 'Lorhana', 'Erick', 'Igor']
print(alunos)
alunos.sort()
print(alunos)

def linhas():
  print('==' * 20)

linhas()
print('PRIMEIRA FORMA')
for i in range(0, len(alunos)):
  print(f'Aluno {alunos[i]} na posição {i}') # print('Alunos {} na posição {} .format(alunos[i], i))

linhas()
print('SEGUNDA FORMA')
for nome in alunos: # nome recebe o VALOR da LISTA
  print(f'Aluno {nome}')

linhas()
print('TERCEIRA FORMA')
for i, nome in enumerate(alunos): # enumerate gera DOIS VALORES -> INDICE e VALOR DA LISTA
  print(f'Aluno {alunos[i]} na posição {i}')  # print('Alunos {} na posição {} .format(alunos[i], i))
```

```python
def inputNum():
  return int(input('Digite um número: '))

numUsu = [] # Lita Vazia
for i in range(0, 5):
  numUsu.append(inputNum())

print(numUsu)
print(f' Maior número  {max(numUsu)}')
print(f' Menor número  {min(numUsu)}')
print(f' Somatório  {sum(numUsu)}')
print(f' Quantidade de itens da lista {len(numUsu)}')
numUsu.sort()
print(numUsu)
```
