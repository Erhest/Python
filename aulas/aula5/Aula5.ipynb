```python
# Faça um jogo da sorte que guarde um numero (0 - 99) e
#  peça para o usuário tentar acertar o numero, exiba se errou pra mais ou menos
# Facil -> 0 -> 25
# Medio -> 0 -> 100
# Dificil -> 0 -> 100 -> Cara -> nâo dou dica
# 4 ou 4 -> Sair

from random import randint

continuar = True
segredo = 0
chances = 0
acertou = False

while(continuar):
  menu = input('**** Jogo da Sorte **** \n 1 - Fácil \n 2 - Médio \n 3 - Díficil \n 4 - Sair \n R: ')
  if(menu.isnumeric()):
    menu = int((menu))
    if(menu > 0 and menu < 4):
      print('Inicio da Partida')
      acertou = False
      if(menu == 1):
        print('**** Fácil ****')
        segredo = randint(0, 25)
        while(not acertou ): # == False
          chute = int(input('Digite um número entre 0 e 25: '))
          chances += 1
          if(chute == segredo):
            print('**** FIM DE PARTIDA')
            print('Você acertou: ', segredo)
            print('Com {} tentativas' .format(chances))
            acertou = True
          elif(chute > segredo):
            print('Menos')
          else:
            print('Mais')
      elif(menu == 2):
        print('**** Médio ****')
        segredo = randint(0, 100)
        while(not acertou ): # == False
          chute = int(input('Digite um número entre 0 e 100: '))
          chances += 1
          if(chute == segredo):
            print('**** FIM DE PARTIDA')
            print('Você acertou: ', segredo)
            print('Com {} tentativas' .format(chances))
            acertou = True
          elif(chute > segredo):
            print('Menos')
          else:
            print('Mais')
      else:
        print('**** Díficil ****')
        segredo = randint(0, 100)
        while(not acertou ): # == False
          chute = int(input('Digite um número entre 0 e 100: '))
          chances += 1
          if(chute == segredo):
            print('**** FIM DE PARTIDA')
            print('Você acerto: ', segredo)
            print('Com {} tentativas' .format(chances))
            acertou = True
          else:
            sera = randint(0, 1)
            if(sera == 0):
              print('Cara')
              if(chute > segredo):
                print('Menos')
              else:
                print('Mais')
            else: 
              print('Coroa')
              print('NÃO POSSO TE AJUDAR')          
    else:
      print('Fim de Jogo')
      continuar = False
  else: 
    print('Coloque um valor VÁLIDO!!')
```
