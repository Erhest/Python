```python
# Função com retorno
from random import randint

def retornaNumeroAleatorio():
  return randint(0, 100)

def mostraIdadeNome(id, nome):
  print('{} tem {} anos' .format(nome, id))
  return id

segredo = retornaNumeroAleatorio()
print('RETORNO: ', segredo)

idade = mostraIdadeNome(27, 'Kenji')
print('RETORNO: ', idade)

```

```python
# Faça um calculadora com (+ - * /) e repita enquanto o usuário quiser

conta = 1

def linhas():
  print('--' * 10)

def menu():
  linhas()
  print('Menu Calculadora')
  linhas()
  escolha = int(input(' 1. + \n 2. - \n 3. / \n 4. * \n R: '))
  return escolha

def inputUsuario():
  continuar = False
  while(not continuar):
    val = input('Digite um valor: ')
    if(val.isnumeric()):
      continuar = True
  return float(val)
  
def soma():
  num1 = inputUsuario()
  num2 = inputUsuario()
  print('Soma: ', num1 + num2)
  
def sub():
  num1 = inputUsuario()
  num2 = inputUsuario()
  print('Sub: ', num1 - num2) 
def div():
  num1 = inputUsuario()
  num2 = inputUsuario()
  print('Div: ', num1 / num2)  
def mult():
  num1 = inputUsuario()
  num2 = inputUsuario()
  print('Mult: ', num1 * num2)

  
while(conta < 5):
  conta = menu()
  if(conta == 1):
    soma()
  elif(conta == 2):
    sub()
  elif(conta == 3):
    div()
  elif(conta == 4):
    mult()
  else:
    linhas()
    print('Tchau, Tchau')
    linhas()
```
